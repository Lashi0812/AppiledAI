```python
from sklearn.datasets import load_digits
import matplotlib.pyplot as plt
import numpy as np
```

```python
X, y = load_digits(return_X_y=True)
X = X.T
X.shape
```

```python
k = np.dot(X.T, X)
k.shape
```

```python
m = 1000
```

```python
A = k[:m, :m]
B = k[:m, m:]
C = B.T @ np.linalg.pinv(A) @ B
kh = np.r_[np.c_[A, B],
           np.c_[B.T, C]]

ax = plt.subplot(121)
ax.imshow(k)

ax = plt.subplot(122)
ax.imshow(kh)
```

```python
np.linalg.norm(k-kh)
```
